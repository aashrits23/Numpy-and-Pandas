# Q27) Create a data frame where order date and ship date both are in JAN-2014

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

jan_2014 = (df['Order_Date'].dt.year == 2014) & (df['Order_Date'].dt.month == 1) & \
                (df['Ship_Date'].dt.year == 2014) & (df['Ship_Date'].dt.month == 1)

jan_2014_df = df[jan_2014]

jan_2014_df
```

# Q28) Generate new data frame for every state (hint: dynamic variable creation)
